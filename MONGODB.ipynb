```python
'''MongoDB: A NoSQL Database
MongoDB is a popular NoSQL (Not Only SQL) database that uses a document-oriented data model. 
It stores data in flexible, JSON-like structures called documents.

Non-Relational Databases
Unlike relational databases (SQL), non-relational databases don't rely on a fixed schema. 
They offer flexibility in data storage and retrieval, making them suitable for handling large volumes of unstructured or semi-structured data.

Key characteristics of NoSQL databases:

Schema-less: Data structure can evolve over time.
Scalability: Can handle massive amounts of data.
High performance: Often optimized for speed and handling large datasets.
Flexibility: Can store various data types within a single document.
When to Choose MongoDB Over SQL
MongoDB is preferred over SQL databases in the following scenarios:

Handling large volumes of unstructured or semi-structured data: MongoDB excels at storing and querying data that doesn't fit neatly into rows and columns.
High-performance applications: MongoDB's flexible schema and indexing capabilities can lead to faster query performance.
Rapid development: The schema-less nature of MongoDB allows for agile development and faster time-to-market.
Real-time analytics: MongoDB's ability to handle large datasets efficiently makes it suitable for real-time analytics applications.
Horizontal scaling: MongoDB can easily scale horizontally to handle increasing data volumes and traffic.
Examples of use cases:

Content management systems
Mobile applications
IoT (Internet of Things) data storage
Real-time analytics
Social media platforms'''
```

```python
#Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
```

```python
'''MongoDB sort() Method :-

The sort() method in MongoDB is used to specify the order in which the query returns matching documents from a given collection.
It takes a document as a parameter that contains a field: value pair that defines the sort order of the result set.
The value is 1 or -1 specifying an ascending or descending sort respectively.   

Syntax:
cursor.sort(sort_key_1=sort_order_1, sort_key_2=sort_order_2, ...)

sort_key_1: The field to sort by.
sort_order_1: 1 for ascending, -1 for descending.

Example:

import pymongo

# Assuming you have a collection named 'products' with documents like:
# { "product_id": 1, "price": 200, "name": "Product A" }
# { "product_id": 2, "price": 150, "name": "Product B" }
# { "product_id": 3, "price": 250, "name": "Product C" }

# Connect to MongoDB (replace with your connection details)
client = pymongo.MongoClient("your_mongodb_host", 27017)
db = client["your_database"]
products = db["products"]

# Sort by price in ascending order
sorted_products = products.find().sort("price", 1)
for product in sorted_products:
    print(product)

# Sort by name in descending order
sorted_products = products.find().sort("name", -1)
for product in sorted_products:
    print(product)'''
```

```python
'''
Deleting Documents in MongoDB
MongoDB provides three primary methods to delete data from a collection: delete_one(), delete_many(), and drop().

delete_one()
Purpose: Deletes a single document that matches the given filter.
Usage: When you want to remove a specific document based on certain criteria.

Example:
collection.delete_one({"name": "John Doe"})

delete_many()
Purpose: Deletes multiple documents that match the given filter.
Usage: When you want to remove multiple documents based on certain criteria.

Example:
collection.delete_many({"age": {"$gt": 30}})

drop()
Purpose: Deletes an entire collection.
Usage: When you want to remove all documents and the collection itself.

Example:
collection.drop()'''
```
