```python
'''
Estimation Statistics:-
Estimation statistics involves using sample data to estimate population parameters. 
In other words, it's about making educated guesses about characteristics of a larger group based on information from a 
smaller subset.   

Point Estimate:-
A point estimate is a single value calculated from sample data to estimate a population parameter.   
Example: If we want to estimate the average height of all adults in a country,
we might measure the height of a random sample of 1000 adults and calculate the sample mean. 
This sample mean is a point estimate of the population mean height.   

Interval Estimate:-
An interval estimate provides a range of values likely to contain the population parameter. 
It's accompanied by a confidence level, indicating the probability that the interval contains the true population parameter.   
Example: Instead of just providing a single value for the average height, 
we could calculate a confidence interval (e.g., 95%) that says the true population mean height is likely between 165 cm and 175 cm.'''
```

```python
#Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.
```

```python
'''
import numpy as np

def estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level=0.95):
  """
  Estimates the population mean using a confidence interval.

  Args:
    sample_mean: The mean of the sample.
    sample_std: The standard deviation of the sample.
    sample_size: The size of the sample.
    confidence_level: The desired confidence level for the interval (default: 0.95).

  Returns:
    A tuple containing the lower and upper bounds of the confidence interval.
  """

  from scipy import stats

  # Calculate the margin of error
  alpha = 1 - confidence_level
  t_critical = stats.t.ppf(1 - alpha / 2, df=sample_size - 1)
  margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

  # Calculate the confidence interval
  lower_bound = sample_mean - margin_of_error
  upper_bound = sample_mean + margin_of_error

  return lower_bound, upper_bound

# Example usage:
sample_mean = 50
sample_std = 10
sample_size = 100
confidence_level = 0.95

lower_bound, upper_bound = estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level)
print(f"The population mean is estimated to be between {lower_bound} and {upper_bound} with {confidence_level * 100}% confidence.")

This function calculates the confidence interval for the population mean using the sample mean, standard deviation,
sample size, and a specified confidence level. It uses the t-distribution to calculate the margin of error and then determines 
the confidence interval bounds.'''
```

```python
#Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.
```

```python
'''
import numpy as np
import scipy.stats as stats

def hypothesis_test(sample1, sample2, alpha=0.05):
  """
  Conducts a two-sample t-test to compare the means of two populations.

  Args:
    sample1: A numpy array containing the first sample.
    sample2: A numpy array containing the second sample.
    alpha: The significance level (default: 0.05).

  Returns:
    A tuple containing the t-statistic, p-value, and a boolean indicating whether to reject the null hypothesis.
  """

  # Calculate sample statistics
  n1 = len(sample1)
  n2 = len(sample2)
  mean1 = np.mean(sample1)
  mean2 = np.mean(sample2)
  std1 = np.std(sample1, ddof=1)
  std2 = np.std(sample2, ddof=1)

  # Calculate pooled standard deviation
  sp = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))

  # Calculate t-statistic
  t_statistic = (mean1 - mean2) / (sp * np.sqrt(1/n1 + 1/n2))

  # Calculate degrees of freedom
  df = n1 + n2 - 2

  # Calculate p-value
  p_value = 1 - stats.t.cdf(t_statistic, df=df)

  # Reject null hypothesis if p-value is less than alpha
  reject_null = p_value < alpha

  return t_statistic, p_value, reject_null

# Example usage:
sample1 = np.array([150, 160, 170, 180, 190])
sample2 = np.array([145, 155, 165, 175, 185])
alpha = 0.05

t_stat, p_value, reject_null = hypothesis_test(sample1, sample2, alpha)
print("t-statistic:", t_stat)
print("p-value:", p_value)
print("Reject null hypothesis:", reject_null)

This Python script performs a two-sample t-test to compare the means of two populations.
It calculates the t-statistic, p-value, and determines whether to reject the null hypothesis based on the specified alpha level.'''
```

```python
#Q7: Write down the steps involved in hypothesis testing.
```

```python
'''
Steps Involved in Hypothesis Testing
Hypothesis testing is a systematic procedure for making decisions about a population based on sample data. 

The steps involved are:

1.State the null and alternative hypotheses: Clearly define the null hypothesis (H₀) and the alternative hypothesis (H₁) based on the research question.
2.Set the significance level (α): Determine the level of significance, which is the probability of rejecting the null hypothesis when it is actually true.
  Common values are 0.05 and 0.01.
3.Collect data: Gather a representative sample from the population.
4.Calculate the test statistic: Determine the appropriate test statistic based on the data and the hypothesis being tested (e.g., z-score, t-statistic, chi-square).
5.Determine the critical value or p-value: Compare the calculated test statistic to a critical value or calculate the p-value.
6.Make a decision: Reject or fail to reject the null hypothesis based on the comparison of the test statistic and critical value or the p-value.
7.Interpret the results: Clearly communicate the findings of the hypothesis test in the context of the research question.
```

```python
#Q8. Define p-value and explain its significance in hypothesis testing.
```

```python
'''
P-Value
The p-value is the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming that the null hypothesis is true.   

In simpler terms, it measures how likely it is that the observed data occurred by chance if the null hypothesis were correct.   

Significance of P-value in Hypothesis Testing
The p-value is crucial in deciding whether to reject or fail to reject the null hypothesis.   

If the p-value is less than the significance level (alpha), we reject the null hypothesis. This indicates strong evidence against the null hypothesis.   
If the p-value is greater than the significance level (alpha), we fail to reject the null hypothesis. This means there is insufficient evidence to reject the null hypothesis.   
For example, if the p-value is 0.03 and the significance level is 0.05, we would reject the null hypothesis as 0.03 is less than 0.05 '''
```

```python
'''
Student's t-Distribution
Student's t-distribution is a probability distribution similar to the normal distribution but with heavier tails.
It's used in statistics when the population standard deviation is unknown and the sample size is small.   

When to Use the t-Distribution
The t-distribution is employed in situations where:

Sample size is small: Typically, when the sample size is less than 30, the t-distribution is preferred over the normal distribution.
Population standard deviation is unknown: If the population standard deviation is not available, the sample standard deviation is used as an estimate, necessitating the use of the t-distribution. 

Common applications of the t-distribution include:

Hypothesis testing: To compare population means or determine the significance of regression coefficients.
Confidence interval estimation: To estimate population parameters with a certain level of confidence.

Key differences between t-distribution and normal distribution:

Shape: T-distribution is flatter and has heavier tails compared to the normal distribution.   
Degrees of freedom: T-distribution is characterized by its degrees of freedom, which depend on the sample size. 
As the sample size increases, the t-distribution approaches the normal distribution.'''
```

```python
#Q12: What is t-statistic? State the formula for t-statistic.
```

```python
'''
T-Statistic
The t-statistic is a measure of how many standard errors the sample mean is from the hypothesized population mean.
It is used in hypothesis testing to determine whether there is a significant difference between the sample mean and the population mean.   

Formula for the t-statistic:
t = (x̄ - μ) / (s / √n)
Where:

t is the t-statistic
x̄ is the sample mean
μ is the hypothesized population mean
s is the sample standard deviation   
n is the sample size
The t-statistic follows a t-distribution with n - 1 degrees of freedom. '''
```

```python
'''Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
Estimate the population mean revenue with a 95% confidence interval.'''
```

```python
'''
Understanding the Problem
We have:

Sample size (n) = 50
Sample mean (x̄) = $500
Sample standard deviation (s) = $50
Confidence level = 95%

Solution Steps
Find the critical value (z-score):

For a 95% confidence interval, the z-score is approximately 1.96.
Calculate the standard error:

Standard error (SE) = s / √n = 50 / √50 = $7.07
Calculate the margin of error:

Margin of error = z * SE = 1.96 * 7.07 ≈ $13.86
Construct the confidence interval:

Lower limit = x̄ - margin of error = 500 - 13.86 ≈ $486.14
Upper limit = x̄ + margin of error = 500 + 13.86 ≈ $513.86
Interpretation
We are 95% confident that the population mean daily revenue for the coffee shop lies between $486.14 and $513.86.'''
```

```python
'''Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.'''
```

```python
'''
Understanding the Problem
Hypothesized decrease in blood pressure (μ) = 10 mmHg
Sample size (n) = 100
Sample mean decrease (x̄) = 8 mmHg
Sample standard deviation (s) = 3 mmHg
Significance level (α) = 0.05

Step 1: State the Hypotheses
Null hypothesis (H₀): The drug decreases blood pressure by 10 mmHg. (μ = 10)
Alternative hypothesis (H₁): The drug decreases blood pressure by less than 10 mmHg. (μ < 10)

Step 2: Calculate the Test Statistic
We will use a one-sample t-test because we are testing a single population mean and the population standard deviation is unknown.
The t-statistic formula is:
t = (x̄ - μ) / (s / √n)
Plugging in the values:
t = (8 - 10) / (3 / √100) = -6.67

Step 3: Determine the Critical Value or P-value
Since this is a one-tailed test (we are testing if the decrease is less than 10 mmHg), we will find the critical value for a one-tailed test with α = 0.05 and degrees of freedom (df) = n - 1 = 99.
Using a t-table or statistical software, we find the critical value to be approximately -1.660.
Alternatively, we can calculate the p-value associated with the calculated t-statistic.

Step 4: Make a Decision
Since the calculated t-statistic (-6.67) is less than the critical value (-1.660), we reject the null hypothesis.
Alternatively, the p-value for this test would be extremely small (close to 0), which also leads to rejecting the null hypothesis.
Conclusion
There is sufficient evidence to conclude that the new drug decreases blood pressure by less than 10 mmHg at the 0.05 significance level.'''
```

```python
'''Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
pounds with a significance level of 0.01.'''
```

```python
'''
Understanding the Problem
We're testing whether the mean weight of a product is less than 5 pounds.

Population mean (μ) = 5 pounds
Population standard deviation (σ) = 0.5 pounds
Sample size (n) = 25
Sample mean (x̄) = 4.8 pounds
Significance level (α) = 0.01

Setting up the Hypotheses
Null hypothesis (H₀): μ ≥ 5 (The mean weight is greater than or equal to 5 pounds)
Alternative hypothesis (H₁): μ < 5 (The mean weight is less than 5 pounds)

Calculating the Test Statistic
Since the population standard deviation is known, we can use a z-test.

Standard error (SE) = σ / √n = 0.5 / √25 = 0.1
z-score = (x̄ - μ) / SE = (4.8 - 5) / 0.1 = -2

Finding the Critical Value or P-value
This is a left-tailed test because we're testing if the mean is less than 5 pounds.
For a significance level of 0.01, the critical z-value is approximately -2.33.
Alternatively, we can find the p-value associated with the calculated z-score of -2.
Making a Decision
Since the calculated z-score (-2) is greater than the critical value (-2.33), we fail to reject the null hypothesis.
Alternatively, the p-value associated with a z-score of -2 is greater than 0.01, leading to the same conclusion.

Conclusion
There is not enough evidence to conclude that the true mean weight of the products is less than 5 pounds at the 0.01 significance level.'''
```
