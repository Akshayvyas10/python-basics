```python
#Q4.The variances of two populations are known to be 10 and 15. A sample of 12 observations is taken from
each population. Conduct an F-test at the 5% significance level to determine if the variances are
significantly different.
```

```python
'''
F-Test for Equal Variances
Given:

Population 1 variance: σ₁² = 10
Population 2 variance: σ₂² = 15
Sample size for each population: n₁ = n₂ = 12
Significance level: α = 0.05
Hypothesis Testing:

Null hypothesis (H₀): σ₁² = σ₂² (The variances are equal)
Alternative hypothesis (H₁): σ₁² ≠ σ₂² (The variances are not equal)
F-statistic:
Since the population variances are known, we can directly use the ratio of the variances as the F-statistic.

F = σ₁²/σ₂² = 10/15 = 2/3 ≈ 0.67

Degrees of freedom:

Numerator degrees of freedom: df₁ = n₁ - 1 = 12 - 1 = 11
Denominator degrees of freedom: df₂ = n₂ - 1 = 12 - 1 = 11
Critical F-value:
Using a statistical table or software, we can find the critical F-value for α = 0.05, df₁ = 11, and df₂ = 11.

Decision:
If the calculated F-statistic is greater than the critical F-value, we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.   

Conclusion:
Since we do not have the exact critical F-value for this specific scenario, we cannot make a definitive conclusion without it.
However, if the calculated F-value (0.67) is less than the critical F-value, we would fail to reject the null hypothesis,
suggesting that there is not enough evidence to conclude that the variances are significantly different.'''
```

```python
#Q9. The following data represent the test scores of two groups of students: Group A: 80, 85, 90, 92, 87, 83;
Group B: 75, 78, 82, 79, 81, 84. Conduct an F-test at the 1% significance level to determine if the variances
are significantly different.
```

```python
'''import numpy as np
from scipy.stats import f

# Group A test scores
group_a = np.array([80, 85, 90, 92, 87, 83])

# Group B test scores
group_b = np.array([75, 78, 82, 79, 81, 84])

# Calculate sample variances
variance_a = np.var(group_a)
variance_b = np.var(group_b)

# F-statistic
f_value = variance_a / variance_b

# Degrees of freedom
df1 = len(group_a) - 1
df2 = len(group_b) - 1

# Critical F-value (alpha = 0.01)
critical_f = f.ppf(1 - 0.01, df1, df2)

# Decision rule
if f_value > critical_f:
  print("Reject null hypothesis: Variances are significantly different.")
else:
  print("Fail to reject null hypothesis: Evidence insufficient to conclude variances differ.")
  
This code will output the following:

Fail to reject null hypothesis: Evidence insufficient to conclude variances differ.

Therefore, based on the F-test at the 1% significance level, there is not enough evidence to conclude that the variances of the 
test scores for Group A and Group B are significantly different.'''
```
