```python
'''
One-Tailed vs. Two-Tailed Tests
The primary difference between one-tailed and two-tailed tests lies in the alternative hypothesis.

One-Tailed Test
Alternative hypothesis: Specifies a direction (greater than or less than).
Example: A new drug will increase blood pressure.
Critical region: Located in one tail of the distribution (either left or right).

Two-Tailed Test
Alternative hypothesis: Does not specify a direction (different from).
Example: There is a difference in height between men and women.
Critical region: Located in both tails of the distribution.

Key Differences
Feature	                            One-Tailed Test	                              Two-Tailed Test
Alternative hypothesis	             Directional	                              Non-directional
Critical region                	     One tail	                                  Both tails
Alpha level distribution	         Entire alpha level in one tail	              Alpha level split between two tails '''
```

```python
#Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.
```

```python
'''
Type I and Type II Errors

Type I Error
A Type I error occurs when we reject a null hypothesis that is actually true. 
It's often referred to as a false positive.
Example: Imagine a medical test for a rare disease.
A Type I error would occur if a healthy person is incorrectly diagnosed with the disease.

Type II Error
A Type II error occurs when we fail to reject a null hypothesis that is actually false.
It's often referred to as a false negative.
Example: Continuing with the medical test example,
a Type II error would occur if a sick person is incorrectly diagnosed as healthy.

In summary:
Type I error: Saying something is there when it's not (false positive).
Type II error: Saying something isn't there when it is (false negative).
The probability of a Type I error is denoted by alpha (α) and is equal to the significance level. 
The probability of a Type II error is denoted by beta (β). '''
```

```python
#Q4: Explain Bayes's theorem with an example.
```

```python
'''
Bayes' Theorem
Bayes' theorem is a mathematical formula used to calculate conditional probabilities. 
It helps us update our beliefs about the probability of an event based on new information.

Formula:

P(A|B) = (P(B|A) * P(A)) / P(B)
Where:

P(A|B) is the probability of A given B
P(B|A) is the probability of B given A
P(A) is the probability of A   
P(B) is the probability of B
Example: Medical Testing
Imagine a medical test for a disease.

P(Disease) = 0.01 (1% of the population has the disease)
P(Positive|Disease) = 0.95 (95% chance of a positive test if you have the disease)
P(Positive|No Disease) = 0.05 (5% chance of a false positive)
We want to find P(Disease|Positive), the probability of having the disease given a positive test result.

Using Bayes' theorem:

P(Disease|Positive) = (P(Positive|Disease) * P(Disease)) / P(Positive)   

To find P(Positive), we use the Law of Total Probability:

P(Positive) = P(Positive|Disease) * P(Disease) + P(Positive|No Disease) * P(No Disease)
= (0.95 * 0.01) + (0.05 * 0.99) = 0.0545

Now we can calculate P(Disease|Positive):

P(Disease|Positive) = (0.95 * 0.01) / 0.0545 ≈ 0.174
So, even if you test positive, there's only a 17.4% chance you actually have the disease.
This highlights the importance of understanding false positives and negatives. '''
```

```python
#Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
```

```python
'''
Confidence Interval
A confidence interval is a range of values that is likely to contain the true value of a population parameter with a certain degree of confidence.   

For example, a 95% confidence interval for the mean height of adult women might be 63 to 65 inches.
This means that if we were to repeat the study many times, we would expect 95% of the confidence intervals to contain the true population mean height.   

Calculating a Confidence Interval
Assumptions:

The sample is a simple random sample.
The population is normally distributed (or the sample size is large enough for the Central Limit Theorem to apply).   

Formula:
Confidence Interval = Sample Mean ± (Critical Value * Standard Error)

Where:
Sample mean (x̄) is the average of the sample data.   
Critical value is determined by the desired confidence level (e.g., 1.96 for a 95% confidence level).
Standard error (SE) is the standard deviation of the sample mean (σ/√n).  

Example:
Suppose we want to estimate the average weight of adult males in a city with a 95% confidence interval.
We take a random sample of 100 men and find that the sample mean weight is 180 pounds with a standard deviation of 20 pounds.

Find the critical value: For a 95% confidence level, the z-score is approximately 1.96.   
Calculate the standard error: SE = 20 / √100 = 2
Calculate the margin of error: Margin of error = 1.96 * 2 = 3.92

Calculate the confidence interval:
Lower limit = 180 - 3.92 = 176.08
Upper limit = 180 + 3.92 = 183.92
Therefore, we are 95% confident that the true average weight of adult males in the city is between 176.08 and 183.92 pounds.'''
```

```python
#Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.
```

```python
'''
Bayes' Theorem Example: Medical Testing

Problem:

A certain disease affects 1% of the population. 
A test for the disease has a 95% accuracy rate for people with the disease (true positive) and a 90% accuracy rate for people without the disease (true negative). 
If a person tests positive, what is the probability they actually have the disease?

Solution:

Let's define the events:
A: Person has the disease
B: Person tests positive
We want to find P(A|B), the probability of having the disease given a positive test.

Given information:
P(A) = 0.01 (1% of the population has the disease)
P(B|A) = 0.95 (95% chance of a positive test if you have the disease)
P(B|A') = 0.10 (10% chance of a false positive, or a positive test if you don't have the disease)

First, we need to calculate P(B):
P(B) = P(B|A) * P(A) + P(B|A') * P(A')
P(B) = (0.95 * 0.01) + (0.10 * 0.99) = 0.0995

Now, we can apply Bayes' theorem:
P(A|B) = (P(B|A) * P(A)) / P(B)
P(A|B) = (0.95 * 0.01) / 0.0995 ≈ 0.0955
Therefore, the probability that a person actually has the disease given a positive test result is approximately 9.55% '''
```

```python
#Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
```

```python
'''
Margin of Error
The margin of error is the amount by which a sample statistic (like the sample mean) is allowed to deviate from the true population parameter. 
It's a measure of uncertainty in the estimate.   

How Sample Size Affects Margin of Error
The margin of error is inversely proportional to the square root of the sample size. 
This means that as the sample size increases, the margin of error decreases.
Intuitively, the larger the sample, the more accurate our estimate of the population parameter is likely to be.   

Formula:

Margin of error = Critical value * (Standard deviation / √n)
Where:

n is the sample size

Example: Opinion Polls
Opinion polls often aim to estimate the proportion of people who support a particular candidate or issue. 
To achieve a smaller margin of error and a more precise estimate, pollsters typically survey a large number of people.
For instance, a poll with a sample size of 1000 people will generally have a smaller margin of error than a poll with a sample size of 500, assuming other factors remain constant.'''
```

```python
#Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.
```

```python
'''
Understanding the Problem
We want to determine if the new weight loss drug is effective based on the given sample data.

Sample size (n) = 50
Sample mean weight loss (x̄) = 6 pounds
Sample standard deviation (s) = 2.5 pounds
Significance level (α) = 0.05 (95% confidence level)

Setting Up the Hypotheses
Null hypothesis (H₀): The drug has no effect on weight loss (μ = 0)
Alternative hypothesis (H₁): The drug is effective in reducing weight (μ > 0)

Calculating the Test Statistic
Since we don't know the population standard deviation, we'll use a t-test.

Degrees of freedom (df) = n - 1 = 50 - 1 = 49
t-statistic = (sample mean - population mean) / (sample standard deviation / sqrt(sample size)) = (6 - 0) / (2.5 / sqrt(50)) ≈ 10.69
Determining the Critical Value or P-value
This is a one-tailed test (we're testing if the weight loss is greater than 0).
For a significance level of 0.05 and 49 degrees of freedom, the critical t-value is approximately 1.676.
Alternatively, we can calculate the p-value associated with the calculated t-statistic.

Making a Decision
The calculated t-statistic (10.69) is much greater than the critical value (1.676).
The p-value associated with a t-statistic of 10.69 and 49 degrees of freedom would be extremely small (close to 0).
Therefore, we reject the null hypothesis. '''
```

```python
#Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
confidence interval for the true proportion of people who are satisfied with their job.
```

```python
'''
Understanding the Problem
Sample size (n) = 500
Sample proportion (p̂) = 65% = 0.65
Confidence level = 95%

Calculating the Confidence Interval
1. Find the standard error (SE):
SE = √(p̂ * (1-p̂) / n) = √(0.65 * 0.35 / 500) ≈ 0.0214

2. Find the critical value (z-score):
For a 95% confidence level, the z-score is approximately 1.96.

3. Calculate the margin of error:
Margin of error = z * SE = 1.96 * 0.0214 ≈ 0.042

4. Calculate the confidence interval:
Lower limit = p̂ - margin of error = 0.65 - 0.042 = 0.608
Upper limit = p̂ + margin of error = 0.65 + 0.042 = 0.692

Interpretation
We are 95% confident that the true proportion of people who are satisfied with their current job is between 60.8% and 69.2%.'''
```

```python
'''Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
significant difference in student performance using a t-test with a significance level of 0.01.'''
```

```python
'''
Understanding the Problem
We want to determine if there is a significant difference in student performance between two teaching methods.

Sample A: Mean (M1) = 85, Standard Deviation (SD1) = 6
Sample B: Mean (M2) = 82, Standard Deviation (SD2) = 5
Significance level (α) = 0.01

Setting Up the Hypotheses
Null hypothesis (H₀): There is no difference in mean scores between the two groups (μ1 = μ2)
Alternative hypothesis (H₁): There is a difference in mean scores between the two groups (μ1 ≠ μ2)
Calculating the Test Statistic
Since we don't know the population standard deviations, we'll use a two-sample t-test.

Assumptions:

The two samples are independent.
The populations from which the samples are drawn are normally distributed or the sample sizes are large enough for the Central Limit Theorem to apply.

Calculate the pooled standard deviation (sp):

sp = sqrt(((n1 - 1) * SD1^2 + (n2 - 1) * SD2^2) / (n1 + n2 - 2))
We need the sample sizes (n1 and n2) to calculate sp.

Calculate the t-statistic:

t = (M1 - M2) / (sp * sqrt(1/n1 + 1/n2))
Determining the Critical Value or P-value
This is a two-tailed test because we're testing for a difference in means, not a specific direction.
Find the critical t-value for a two-tailed test with a significance level of 0.01 and the calculated degrees of freedom (df = n1 + n2 - 2).
Alternatively, calculate the p-value associated with the calculated t-statistic.

Making a Decision
Compare the calculated t-statistic with the critical value or compare the p-value to the significance level.
If the t-statistic falls in the rejection region (or the p-value is less than alpha), reject the null hypothesis.
Otherwise, fail to reject the null hypothesis.'''
```

```python
'''Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean
of 65. Calculate the 90% confidence interval for the true population mean.'''
```

```python
'''
Understanding the Problem
We have a population with known parameters:

Population mean (μ) = 60
Population standard deviation (σ) = 8
Sample size (n) = 50
Sample mean (x̄) = 65
Confidence level = 90%

Calculating the Confidence Interval
Since we know the population standard deviation, we can use the z-distribution.

1. Find the z-score for a 90% confidence level:
For a 90% confidence level, the z-score is approximately 1.645.

2. Calculate the standard error:
Standard error (SE) = σ / √n = 8 / √50 ≈ 1.13

3. Calculate the margin of error:
Margin of error = z * SE = 1.645 * 1.13 ≈ 1.86

4. Calculate the confidence interval:
Lower limit = x̄ - margin of error = 65 - 1.86 ≈ 63.14
Upper limit = x̄ + margin of error = 65 + 1.86 ≈ 66.86

Interpretation
We are 90% confident that the true population mean lies between 63.14 and 66.86.'''
```
